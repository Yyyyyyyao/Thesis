```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
new_df = pd.read_csv('xg_2017_02_step2_time1.csv', parse_dates=['Date'], date_parser=dateparse)
```

```python
new_df
```

```python
new_df.insert(4, 'time2', 0) # create new column containing time3
```

```python
new_df
```

```python
# time1
for index, row in new_df.iterrows():
    cur_hour = row['Hours']
    cur_location = row['PULocationID']
    cur_date = row['Date']
    
    if(cur_date == '2017-02-01'):
        if cur_hour == 0:
            new_df.at[index,'time1'] = 0
        
        if cur_hour == 1:
            new_df.at[index,'time1'] = 0
            
        if cur_hour == 2:
            new_df.at[index,'time1'] = 0
    else:
        
        if cur_hour > 2 :
            tar_hour = cur_hour - 3
            tar_date = cur_date
        elif cur_hour == 2:
            tar_hour = 23
            tar_date = cur_date - timedelta(days=1)
        elif cur_hour == 1:
            tar_hour = 22
            tar_date = cur_date - timedelta(days=1)
        elif cur_hour == 0:
            tar_hour = 21
            tar_date = cur_date - timedelta(days=1)
            
        tar_list = new_df.index[(new_df['Hours'] == tar_hour) & (new_df['Date'] == tar_date)& (new_df['PULocationID'] == cur_location)].tolist()
        if not tar_list :
            count = 0
        else:
            i = tar_list[0]
            count = new_df.iloc[i]['count']

        new_df.at[index,'time1'] = count
```

```python
# time 2
for index, row in new_df.iterrows():
    cur_hour = row['Hours']
    cur_location = row['PULocationID']
    cur_date = row['Date']
    
    if(cur_date == '2017-02-01'):
        if cur_hour == 0:
            new_df.at[index,'time2'] = 0
        
        if cur_hour == 1:
            new_df.at[index,'time2'] = 0

    else:
        
        if cur_hour > 1 :
            tar_hour = cur_hour - 2
            tar_date = cur_date
        elif cur_hour == 1:
            tar_hour = 23
            tar_date = cur_date - timedelta(days=1)
        elif cur_hour == 0:
            tar_hour = 22
            tar_date = cur_date - timedelta(days=1)
            
        tar_list = new_df.index[(new_df['Hours'] == tar_hour) & (new_df['Date'] == tar_date)& (new_df['PULocationID'] == cur_location)].tolist()
        if not tar_list :
            count = 0
        else:
            i = tar_list[0]
            count = new_df.iloc[i]['count']

        new_df.at[index,'time2'] = count
```
