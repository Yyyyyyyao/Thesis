```python
import pandas as pd
import numpy as np
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
new_df = pd.read_csv('yellow_tripdata_2017-01.csv', parse_dates=['tpep_pickup_datetime'], date_parser=dateparse)
```

```python
df = new_df[['tpep_pickup_datetime', 'PULocationID']]
df
```

```python
# new_df['Hours'] = new_df['tpep_pickup_datetime'].dt.hour
# new_df['Date'] = new_df['tpep_pickup_datetime'].dt.date
# new_df

df['Hours'] = new_df['tpep_pickup_datetime'].dt.hour
df['Date'] = new_df['tpep_pickup_datetime'].dt.date
df
```

```python
df_1 = new_df.loc[(new_df['PULocationID'] == 1)]
df_1
```

```python
df_group = df_1.groupby(['Date', 'Hours']).count()
```
