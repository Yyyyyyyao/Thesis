```python
import pandas as pd
```

```python
unnom_df = pd.read_csv('../normalized_dataset/xg_2017_0102Combine_UNnorm.csv')
nom_df = pd.read_csv('../normalized_dataset/xg_2017_0102Combine_Normalization.csv')
```

```python
index_43 = 43 / unnom_df['PULocationID'].max()
index_12 = 12 / unnom_df['PULocationID'].max()
index_161 = 161 / unnom_df['PULocationID'].max()
index_2 = 2 / unnom_df['PULocationID'].max()
index_132 = 132 / unnom_df['PULocationID'].max()
index_138 = 138 / unnom_df['PULocationID'].max()
```

```python
location_43 = nom_df.loc[nom_df['PULocationID'] == index_43]
location_12 = nom_df.loc[nom_df['PULocationID'] == index_12]
location_161 = nom_df.loc[nom_df['PULocationID'] == index_161]
location_2 = nom_df.loc[nom_df['PULocationID'] == index_2]
location_132 = nom_df.loc[nom_df['PULocationID'] == index_132]
location_138 = nom_df.loc[nom_df['PULocationID'] == index_138]
```

```python
location_43.to_csv('normalized_location_43.csv', index=False)
location_12.to_csv('normalized_location_12.csv', index=False)
location_2.to_csv('normalized_location_2.csv', index=False)
location_161.to_csv('normalized_location_161.csv', index=False)
location_132.to_csv('normalized_location_132.csv', index=False)
location_138.to_csv('normalized_location_138.csv', index=False)
```
